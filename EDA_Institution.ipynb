```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('Guardian_Dataset.xlsx', sheet_name=None)
```

```python
print(data.keys())
```

```python
df_insti = data['Institution Level Data']
```

```python
df_insti.head()
```

```python
df_insti.info()
```

```python
df_insti.isnull().sum()
```

```python
df_insti_sorted = df_insti.sort_values(by=['Year', 'Rank'])
df_insti_sorted
```

```python
df_insti_sorted.duplicated().sum()
```

```python
df_insti_sorted[df_insti_sorted['Institution'] == 'Queen Margaret']
```

```python
df_insti_sorted[df_insti_sorted.apply(lambda row: any(str(cell).isspace() for cell in row), axis=1)]
```

```python
df_insti_sorted_cleaned = df_insti_sorted.replace(r'^\s*$', np.nan, regex=True)
```

```python
columns_to_convert = ['Satisfied with Course', 'Satisfied with Teaching', 'Satisfied with Feedback', 
                      'Student to Staff Ratio', 'Spend per Student', 'Value Added Score']
df_insti_sorted_cleaned[columns_to_convert] = df_insti_sorted_cleaned[columns_to_convert].astype(float)
```

```python
df_insti_sorted_cleaned.info()
```

```python
df_cleaned = df_insti_sorted_cleaned.fillna(method='bfill')
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned
```

```python
df_cleaned[df_cleaned['Institution'] == 'Queen Margaret']
```

```python
df_cleaned[df_cleaned.apply(lambda row: any(str(cell).isspace() for cell in row), axis=1)]
```

```python
df_cleaned.to_csv('Institution_cleaned.csv', index=False)
```

```python
df = pd.read_csv('Institution_cleaned.csv')
```

```python
df_uwe = df[df['Institution'] == 'UWE Bristol']
df_uwe
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_uwe[['Satisfied with Course', 'Satisfied with Teaching', 'Satisfied with Feedback']])
plt.title('Trends over Time for Satisfaction')
plt.xlabel('Year')
plt.ylabel('Satisfaction Level')
plt.show()
```

```python
corr_matrix = df_uwe.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(x='Institution', y='Career after 15 months', data=df_uwe)
plt.title('Career after 15 months by Institution')
plt.show()
```
